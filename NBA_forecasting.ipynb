## Formatting DataFrames -- Moddelling -- Forecast

```python
sch_reg = matches_schedule.copy()
stat_reg = teams_stats.copy()
stat_reg.replace(to_replace=[','], value=['.'], regex=True, inplace=True)


k = 0
for i in stat_reg['off_FT_FGA__s']:
    stat_reg['off_FT_FGA__s'][k] = '0' + stat_reg['off_FT_FGA__s'][k]
    stat_reg['FTr__s'][k] = '0' + stat_reg['FTr__s'][k]
    stat_reg['tree_PAr__s'][k] = '0' + stat_reg['tree_PAr__s'][k]
    stat_reg['TS__s'][k] = '0' + stat_reg['TS__s'][k]
    stat_reg['off_eFG__s'][k] = '0' + stat_reg['off_eFG__s'][k]
    stat_reg['deff_eFG__s'][k] = '0' + stat_reg['deff_eFG__s'][k]
    stat_reg['off_FT_FGA__s'][k] = '0' + stat_reg['off_FT_FGA__s'][k]
    stat_reg['deff_FT_FGA__s'][k] = '0' + stat_reg['deff_FT_FGA__s'][k]
    k += 1
#pd.set_option('display.max_columns', None)
for i in range(0, len(stat_reg.columns)):
    stat_reg.iloc[:,i] = pd.to_numeric(stat_reg.iloc[:,i], errors='ignore')
for x in range(0, len(sch_reg.columns)):
    sch_reg.iloc[:,x] = pd.to_numeric(sch_reg.iloc[:,x], errors='ignore')
```

```python
predicted_matches = []
exist = 0
for mac_sıra in range(12):

    selected = 0
    for i in sch_reg.iloc[:,3]:
        if pd.isna(i):
            break
        selected += 1

    selected += mac_sıra
    #print(matches_schedule.iloc[selected])


    reg_df = stat_reg.copy()
    reg_df.drop(['Team__s', 'Arena__s', 'Attend__s', 'Attend_G__s', 'Age__s', 'W__s', 'L__s', 'PW__s', 'PL__s', 'SRS__s', 'NRtg__s', 'MOV__s', 'SOS__s', 'ORtg__s','DRtg__s','Pace__s','FTr__s','tree_PAr__s','TS__s'], axis=1, inplace=True)

    reg_1 = reg_df.copy()
    reg_1.drop(reg_1.index, inplace=True)

    reg_forecast = reg_df.copy()
    reg_forecast.drop(reg_forecast.index, inplace=True)

    if exist == 0:
        all_reg_forecast = reg_forecast.copy()
        exist += 1
    
    t1 = 0
    for i in stat_reg['Team__s']:
        if i == matches_schedule.iloc[selected,2]:
            break
        t1 += 1
    t2 = 0
    for i in stat_reg['Team__s']:
        if i == matches_schedule.iloc[selected,4]:
            break
        t2 += 1
    selected_match_data = (reg_df.iloc[t1,:]) - (reg_df.iloc[t2,:])
    reg_forecast = reg_forecast.append(selected_match_data, ignore_index=True)


    v = selected-mac_sıra
    try:
        sayac = 0
        for i in range(len(sch_reg)):
            t1 = 0
            for i in stat_reg['Team__s']:
                if i == matches_schedule.iloc[v-1 ,2]:
                    break
                t1 += 1
            t2 = 0
            for i in stat_reg['Team__s']:
                if i == matches_schedule.iloc[v-1 ,4]:
                    break
                t2 += 1

            data_var = (reg_df.iloc[t1,:]) - (reg_df.iloc[t2,:])
            data_score = int(matches_schedule.iloc[v-1 ,3]) + int(matches_schedule.iloc[v-1 ,5])

            reg_1 = reg_1.append(data_var, ignore_index = True)
            reg_1.loc[sayac,'Score'] = data_score

            v -= 1
            sayac += 1
    except:
        None


    x = reg_1.iloc[:, :-1]
    y = reg_1.iloc[:, -1]

    model = LinearRegression().fit(x,y)


    all_reg_forecast = all_reg_forecast.append(reg_forecast)

    predict1 = model.predict(np.array([reg_forecast.iloc[0,:]]))[0]
    #predict2 = model.predict(np.array([reg_forecast.iloc[1,:]]))[0]

    predicted_matches.append(f"|{matches_schedule.iloc[selected,2]} - {matches_schedule.iloc[selected,4]}|  > {predict1}")
    #print(predict2, matches_schedule.iloc[selected,4])
```

```python
print(f'Score: {model.score(x, y)}')
print(f'Coefficient: {model.coef_}')
print(f'Intercept: {model.intercept_}')
```
